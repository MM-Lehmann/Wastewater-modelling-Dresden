```python
import pandas as pd
from joblib import Memory
from datetime import datetime
memory = Memory(f"cachedir/{datetime.today().date()}", compress=True)

@memory.cache
def cached_read(url):
    return pd.read_csv(url)
```

```python
df = cached_read("https://github.com/robert-koch-institut/SARS-CoV-2_Infektionen_in_Deutschland/blob/master/Aktuell_Deutschland_SarsCov2_Infektionen.csv?raw=true")
df = df[df.IdLandkreis == 14612]
df
```

```python
group = df.groupby(["Refdatum", "Altersgruppe"]).sum().filter(("AnzahlFall",)).interpolate().reset_index("Altersgruppe")
# Demographics
group.loc[group.Altersgruppe == "A00-A04", "AnzahlFall"] /=  .27400
group.loc[group.Altersgruppe == "A05-A14", "AnzahlFall"] /=  .54400
group.loc[group.Altersgruppe == "A15-A34", "AnzahlFall"] /= 1.43200
group.loc[group.Altersgruppe == "A35-A59", "AnzahlFall"] /= 1.82600
group.loc[group.Altersgruppe == "A60-A79", "AnzahlFall"] /= 1.07350
group.loc[group.Altersgruppe == "A80+",    "AnzahlFall"] /=  .46100

df_out = group.groupby("Altersgruppe").rolling(7, min_periods=1).sum().rename(columns={"AnzahlFall": "rki_7tage"})
df_out["rki_14tage"] = group.groupby("Altersgruppe").rolling(14, min_periods=1).sum()

# df_out = df_out / 5.61
df_out = df_out.pivot_table(index="Refdatum", columns="Altersgruppe")
df_out.index = pd.to_datetime(df_out.index)
df_out = df_out.interpolate("time", axis=0).fillna(0)
df_out.columns = [' '.join(col).strip() for col in df_out.columns.values]
df_out.plot(figsize=(25,6), title="Refdatum")
df_out
```

```python
df_out.to_csv("Dresden_RKI_data_altersgruppen.csv")
```
