```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv(
    "~/user/Downloads/Indicadors_de_pr_sencia_de_rastres_gen_tics_del_virus_SARS-CoV-2_a_les_aig_es_residuals.csv"
)
df["date"] = pd.to_datetime(df["id mostra"].str.split("-", n=1).str[1])
df[::200]
```

```python
df_filtered = df.groupby("depuradora").filter(lambda x: len(x) > 70)
print(df["depuradora"].unique())
len(df_filtered)
```

```python
df_filtered.groupby("date").sum().reset_index().plot("date", "N2(CG/L)")
```
